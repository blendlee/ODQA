```python
import json
from collections import defaultdict
```

```python
files = [
 'new_kluebert_2.json',
 'new_koelectra3.json',
 #'org_kluebert2_56.json',
 'org_rl2_68.json',
 'new_rl_2.json',
 'rlcnn2_60.json',
 'org_rb2_57.json',
 #'org_koelectra3_58.json'

] # 불러올 파일들 경로

 
json_files = []
for file in files:
    with open(f"prediction_json/{file}") as f:
        json_file = json.load(f)
        json_files.append(json_file)
```

```python
result = dict()
    
for key in json_files[0].keys(): 
    polls = defaultdict(int)
    
    for json_file in json_files:
        predict = json_file[key]
        polls[predict] += 1
    
    sorted_polls = sorted(polls.items(), key=lambda x: x[1], reverse=True)
    print(sorted_polls)
    voted = sorted_polls[0][0]

    result[key] = voted
```

```python
final = 'org+_new_ensemble.json'
with open(final, 'w') as file:
    json.dump(result, file,ensure_ascii = False,indent="\t")
```
