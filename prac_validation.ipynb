```python
import json
from datasets import load_from_disk
```

```python
with open('/opt/ml/input/code/nbest_predictions_koelectra.json') as f:
    nbest_pred_koelectra = json.load(f)
with open('/opt/ml/input/code/nbest_predictions_robertalarge2.json') as f:
    nbest_pred_rl = json.load(f)
```

```python
val_dataset=load_from_disk('/opt/ml/input/data/train_dataset')['validation']
```

```python
score_with_npred=0
score_with_1pred=0
score_with_2pred=0
score_with_3pred=0
score_with_4pred=0
koelectra=[]
for id,answers in zip(val_dataset['id'],val_dataset['answers']):
    found=False
    answer = answers['text'][0]
    for rank, candiate in enumerate(nbest_pred_koelectra[id]):
        if candiate['text'] == answer:
            if rank ==0:
                score_with_1pred+=1
            elif rank ==1:
                score_with_2pred+=1
            elif rank==2:
                score_with_3pred+=1
            elif rank ==3:
                score_with_4pred+=1
            print(rank+1,answer)
            koelectra.append((rank+1,answer))
            score_with_npred+=1
            found=True
            break
    if not found:
        print(answer,nbest_pred_koelectra[id])
            
```
